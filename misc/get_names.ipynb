```python
import glob
from pathlib import Path
import json
```

```python
plant_paths = glob.glob("plant_mode/*")
plant_paths = [Path(pp) for pp in plant_paths]
plant_paths
```

```python
import re

def parse_plant_filename(filename):
    pattern = r'^([A-Za-z]+)(?:_([A-Za-z]+))?(?:_\(([^)]+)\))?(?:_(\d+))?$'
    match = re.match(pattern, filename)
    if match:
        genus = match.group(1)
        species = match.group(2)
        common_name_raw = match.group(3)
        number_raw = match.group(4)

        scientific_name = f"{genus} {species}" if species else genus
        common_name = common_name_raw.replace('_', ' ') if common_name_raw else None
        number = int(number_raw) if number_raw else None

        return scientific_name, common_name, number
    else:
        return None
```

```python
plant_dict = {}
for path in plant_paths:
    filename = path.name.replace(path.suffix, '')
    print(filename)
    scientific_name, common_name, number = parse_plant_filename(filename)
    print(scientific_name, common_name, number)
    print()
    plant_dict[str(path)] = (scientific_name, common_name)
with open('plant_dict.json', 'w') as f:
    json.dump(plant_dict, f)
```
